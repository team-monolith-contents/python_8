함수는 입력값을 받아 처리하고, 결과를 반환하는 코드의 덩어리입니다.   

**파이썬의 함수와 수학의 함수**
* 파이썬의 함수와 수학의 함수는 목적과 구현 방식에서 차이가 있지만, 모두 입력값을 받아 일정한 수식으로 처리한 결과를 반환한다는 공통점이 있습니다.
* 수학에서 함수를 공부할 때 배웠던 일차함수를 생각해봅시다. 
> 예를 들어 `f(x) = 7x+3`라는 함수의 x에 3을 넣으면, 결과 7*3+3=24를 반환해줍니다.
* 즉 입력에 대해 일정한 수식을 적용하여 결과를 반환합니다.  
* 이와 같이, 컴퓨터의 함수도 입력값에 대해 정의된 기능을 처리하는 일을 합니다.   





**함수의 필요성**
* 함수는 재사용이 가능한 코드로, 코드를 재사용하고 중복을 제거할 수 있습니다. 
* 프로그램의 구조를 단순하게 만들어줍니다. 
* 코드의 가독성을 높이고 유지보수성을 개선시킵니다.

### 함수 사용하기

함수는 `함수이름()`의 방법으로 사용할 수 있습니다.<br>
지금까지 우리가 사용해온 `print()`나 `input()` 모두 함수이기 때문에 같은 방법으로 사용합니다.<br>

`input()`과 같은 함수는 괄호안에 아무것도 넣지 않고 사용할 수도 있지만 `len()`과 같은 몇몇 함수들은 괄호안에 특정 값들을 넣어주어야 합니다.<br>
이 때 괄호안에 들어가는 값을 인자(argument)라고 합니다.<br>
매개변수는 함수를 사용할 때 함수에 전달되는 값이나 변수를 의미합니다.<br>
해당 인자들은 함수 안에서 재사용 될 수 있습니다.<br>

`len()` 함수를 예로 들면 `len([1,2,3])`로 리스트 `[1,2,3]`의 길이를 구할 수 있습니다.<br>
`len()` 함수 내부에서는 리스트 `[1,2,3]`의 길이를 구하기 위해서 연산을 할텐데 이를 위해 `[1,2,3]` 리스트를 알고 있어야 합니다.<br>
우리는 인자로 리스트 `[1,2,3]`을 전달함으로써 len() 함수에게 리스트 `[1,2,3]`을 알려줄 수 있습니다.<br>

`len()`과 같은 함수들은 결과값을 반환합니다.<br>

```python
a = print("abc") # 반환값이 없는 함수
b = len("abc") # 반환값이 있는 함수
print(a,b)
```
함수의 반환값은 변수 처럼 사용할 수 있습니다.<br>
함수의 반환값은 인자에 따라 바뀝니다.

### 반환값이 없는 함수 만들기

파이썬에서 함수는 `def` 키워드를 사용하여 정의합니다.

함수의 기본 구조는 다음과 같습니다.

```python
def function_name(parameters):
    # 함수의 동작을 구현하는 코드
    # 파라미터(parameters)는 함수의 입력값으로 사용됩니다.
    # 함수의 반환값이 있는 경우, return 키워드를 사용하여 반환합니다.
    # 들여쓰기를 하여 함수 코드를 구분합니다.
```

- **function_name:** 함수의 이름으로, 함수를 호출할 때 사용됩니다.
- **parameters:** 함수의 매개변수로, 함수가 입력값을 받을 때 사용됩니다. 매개변수는 필수가 아니며, 없을 수도 있습니다.

아래의 예시와 함께 설명해보겠습니다.

```python
def hello(name):
    print("Hello, ", name)
    
hello("codle")
```

위의 코드는 `hello` 함수를 정의하고 사용하는 코드입니다.<br>
`hello` 함수는 `name`이라는 매개변수를 사용합니다.<br>
함수는 이름을 입력받아 "Hello, {name}"을 출력합니다. name의 자리에는 전달받은 매개변수가 들어갑니다.<br>
`hello("codle")`의 코드로 함수를 호출하면 "Hello, codle"을 출력할 것입니다.<br>

**예제 4 : 매개변수 없는 함수 만들기**

````python
# 예제_4_매개변수_없는_함수_만들기

def hello():
    print("Hello, codle")
    
hello()
hello()
````

**예제 5 : 매개변수가 하나인 함수 만들기**

````python
# 예제_5_매개변수가_하나인_함수_만들기

def hello(name):
    print("Hello, My name is",name)
    
hello("코들")
hello("codle")
````

### 반환값이 있는 함수 만들기

함수는 `return` 키워드를 사용해 값을 반환할 수 있습니다.<br>
함수가 실행되다가 `return` 키워드를 만나면 함수가 즉시 종료되고 함수가 호출되었던 바로 다음 줄 부터 코드가 실행됩니다.<br>

`return` 뒤에 특정 데이터를 넣어주면 함수가 해당 값을 반환합니다. 

아래의 예시와 함께 설명해보겠습니다.

```python
def add(a, b):
    print(a, b) # 1
    return a+b # 2
    print(a+b) # 3
    
result = add(3, 5) 
```

위의 코드는 `add` 함수를 정의하고 사용하는 코드입니다.<br>
`add` 함수는 `a`와 `b`를 매개변수로 사용합니다.<br>
`add` 함수가 호출되면 함수는 1에서 3 순으로 실행이 됩니다.<br>
우선 1에서 a와 b가 출력이 됩니다.<br>
그 다음 함수는 `return a+b`를 만나게 됩니다.<br>
이 코드는 이 함수를 `a+b` 값을 반환하면서 종료하라는 뜻입니다.<br>
따라서 이 함수의 3은 실행되지 않고 a+b 값을 반환하면서 종료되게 됩니다.<br>
위의 코드에서는 add 함수의 인자로 a = 3, b = 5가 주어졌으므로 3+5=8의 값을 반환하면서 함수가 종료되게됩니다.<br>
반환된 값 8은 result 라는 변수에 저장되게 됩니다.

**예제 6 : 두 수의 곱을 반환하는 함수 작성하기**

````python
# 예제_6_두_수의_곱을_반환하는_함수_작성하기

def product(a,b):
    return a*b

print(product(3,5))
````

````python
# 정답코드

import random

# 정답 생성 함수
def make_answer():
    while True:
        answer = random.randint(100,999)
        answer = str(answer)
        if answer[0] != answer[1] and answer[0] != answer[2] and answer[1] != answer[2]:
            return answer
        
# 볼, 스트라이크 판별 함수
def strike_ball(player, answer):
    s = 0
    b = 0
    for j in range(len(player)):
        if player[j] == answer[j]:
            s += 1
        elif player[j] in answer:
            b += 1
    return s, b

# 정답 판정 함수
def judge(s, b):
    if s == 3:
        return True
    else:
        return False
    
# 기록 추가 함수
def add_history(history, result, player, s,b):
    history.append(player)
    result.append([s,b])
    return history, result
````
